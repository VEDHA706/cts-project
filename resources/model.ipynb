```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"E:\Projects\CTS\data\data.csv")
```

```python
data.head()
```

## 1. Meal Type and Dish Category Distribution
*Pie Chart:* To show the proportion of different meal types (Breakfast, Lunch, Dinner) and dish categories (Vegetarian, Non-Vegetarian).

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
MealType = data['MealType'].to_list()
DishCategory = data['DishCategory'].to_list()

# Extract unique categories and meal types
categories = list(set(DishCategory))
MealTypes = list(set(MealType))

# Count the occurrences in each category
category_counts = [DishCategory.count(category) for category in categories]
mealtype_counts = [MealType.count(mealtype) for mealtype in MealTypes]

# Plotting
plt.figure(figsize=(10, 5))

# First plot for DishCategory
plt.subplot(1, 2, 1)
plt.pie(category_counts, labels=categories, colors=['gray', 'teal'], autopct='%1.1f%%')
plt.title('Distribution of NVeg and Veg')

# Second plot for Meal Type
plt.subplot(1, 2, 2)
plt.pie(mealtype_counts, labels=MealTypes, colors=['lightblue', 'cyan', 'pink'], autopct='%1.1f%%')
plt.title('Distribution of Breakfast, Dinner, and Lunch')

print("Data Balance:")
# Display the plots
plt.tight_layout()
plt.show()
```

## 2. Distribution of Nutrients
*Histogram or Box Plot:* To visualize the distribution of calories, proteins, fats, carbs, etc

```python
# Create a figure with subplots
plt.figure(figsize=(10, 5))

# First subplot: Histogram of Calories
plt.subplot(1, 2, 1)
sns.histplot(data['Calories'], kde=True, color='skyblue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')

# Second subplot: Box Plot for Nutritional Values
plt.subplot(1, 2, 2)
sns.boxplot(data=data[['Calories', 'Proteins', 'Fat', 'Carbs']], palette="Set2")
plt.title('Nutritional Values of Meals')

# Show both plots together
plt.tight_layout()  # Adjust the layout to prevent overlap
plt.show()
```

## 3. Scatter Plot for Caloric Content vs. Other Nutrients
Scatter Plot: To see the relationship between calories and other nutrients like proteins, fat, etc.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Calories', y='Proteins', data=data, hue='DishCategory', style='MealType', palette='deep')
plt.title('Calories vs. Proteins')
plt.xlabel('Calories')
plt.ylabel('Proteins')
plt.show()
```

## 4. Top Dishes by Calories
*Bar Plot:* To visualize the top N dishes with the highest calorie content.

```python
top_dishes = data[['DishName', 'Calories']].sort_values(by='Calories', ascending=True).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='Calories', y='DishName', data=top_dishes, palette='viridis',hue=top_dishes['DishName'])
plt.title('Top 10 Dishes by Calories')
plt.xlabel('Calories')
plt.ylabel('Dish Name')
plt.show()
```
