```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"E:\Projects\CTS\data\data.csv")
```

```python
data.head()
```

```python
MealType = data['MealType'].to_list()
DishCategory = data['DishCategory'].to_list()
```

```python
# Count the elements in each list
count_MealType = len(MealType)
count_DishCategory = len(DishCategory)

# Data for plotting
categories = ['List 1', 'List 2']
counts = [count_MealType, count_DishCategory]

# Plotting
plt.figure(figsize=(8, 5))

# First plot for List 1
plt.subplot(1, 2, 1)
plt.bar(categories[0], counts[0], color='blue')
plt.title('Count in List 1')
plt.ylabel('Count')

# Second plot for List 2
plt.subplot(1, 2, 2)
plt.bar(categories[1], counts[1], color='green')
plt.title('Count in List 2')

# Display the plots
plt.tight_layout()
plt.show()
```
